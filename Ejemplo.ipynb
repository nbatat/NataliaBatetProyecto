```python
#seleccionando un fila
fila_random = datos.loc[0]
print(fila_random)
```

```python
#seleccionando una columna
columna_random = datos.loc[:,"Area covered"]
print(columna_random)
```

```python
#.loc[]: Se utiliza para la selección basada en etiquetas. Esto significa que puedes seleccionar filas o columnas utilizando los índices de fila y columna específicos (etiquetas)
#en el DataFrame. Cuando utilizas .loc[], Pandas busca las etiquetas específicas que especificas.

#.iloc[]: Se utiliza para la selección basada en posiciones enteras de fila y columna. Esto significa que puedes seleccionar filas o columnas utilizando los índices enteros de fila
#y columna en el DataFrame. Cuando utilizas .iloc[], Pandas selecciona las filas o columnas en función de sus posiciones en el DataFrame.
```

```python
columna_random_2 = datos.iloc[:,5]
print(columna_random_2)
```

```python
fila_random_2 = datos.iloc[9]
print(fila_random_2)
```
